```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# grafo plain
G = nx.Graph()
G
```

```python
# asignar nodos
G.add_node('a')
G.add_node('b')
G.add_node('c')
G.add_node('d')
G.add_node('e')

# agregar varios nodos
lista_de_nodos = ['f', 'g', 'h', 'i', 'j']
G.add_nodes_from(lista_de_nodos)

# ver nodos
G.nodes()
```

```python
# agregar aristas
G.add_edge('a', 'b')
G.add_edge('a', 'c')
G.add_edge('a', 'd')
G.add_edge('b', 'e')
G.add_edge('c', 'e')
G.add_edge('d', 'e')
G.add_edge('e', 'f')
G.add_edge('f', 'g')
G.add_edge('g', 'h')
G.add_edge('h', 'i')
G.add_edge('i', 'j')


# agregar varias aristas
lista_de_aristas = [('a', 'e'), ('b', 'c'), ('d', 'f'), ('g', 'h')]
G.add_edges_from(lista_de_aristas)

# ver aristas
G.edges()
```

```python
# dibujar el grafo
nx.draw(G, with_labels=True)
plt.show()
```

```python
# customizar el dibujo
pos = nx.spring_layout(G)  # posiciones para todos los nodos
nx.draw_networkx_nodes(G, pos, node_size=700)  # nodos
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), width=6)  # aristas
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')  # etiquetas
plt.axis('off')  # no mostrar ejes
plt.show()  # mostrar el dibujo
```

```python
nx.draw_networkx(G,
        with_labels=True,
        node_color='blue',
        node_size=1600,
        font_color='white',
        font_size=16,
        )
```

```python
# iteradores

print("Nodos:")
for node in G.nodes():
    print(node)

print("Aristas:")
for edge in G.edges():
    print(edge)
```

```python
print("Número de Nodos:", G.number_of_nodes())
print("Número de Aristas:", G.number_of_edges())
```

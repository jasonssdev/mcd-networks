```python
import networkx as nx
from matplotlib import pyplot as plt
```

```python
G = nx.petersen_graph()
# Dibuja el grafo
nx.draw_networkx(G, with_labels=True)
plt.show()
```

```python
G_directed = nx.gn_graph(10, seed = 1)
nx.draw_networkx(G_directed, with_labels=True)
plt.show()
```

```python
print(len(list(G.neighbors(8))))
```

```python
print(len(list(G_directed.neighbors(8))))
```

```python
print("Grados del Nodo 0:",G.degree(0))
print("Grados del Nodo 0:",G_directed.degree(0))
```

```python
print("Nodos entrantes:", G_directed.in_degree(2))
print("Nodos salientes:",G_directed.out_degree(2))
# Si Utilizamos degree
print("Grados Totales:",G_directed.degree(2))
```

```python
# densidad

print("Densidad del grafo no dirigido:", nx.density(G))
print("Densidad del grafo dirigido:", nx.density(G_directed))
```

```python
# subgraph
G_sub = G.subgraph([0,1,2,3,4,5])
nx.draw_networkx(G_sub, with_labels=True)
plt.show()
```

```python
# matriz de adyacencia
A = nx.adjacency_matrix(G).todense()
print(A)
```

```python
A = nx.adjacency_matrix(G_sub).todense()
print(A)
```

```python
A = nx.adjacency_matrix(G_directed).todense()
print(A)
```

```python
#Guardar lista de adyacencia
nx.write_adjlist(G, path="graphs/A_list_path")

# Crear un grafo desde una lista de adyacencia
G_new = nx.read_adjlist("./graphs/A_list_path")
nx.draw_networkx(G_new, with_labels=True)
plt.show()
```
